```python
import pandas as pd

pd.set_option('max_colwidth', None)
pd.set_option('display.max_columns', 37)
```

```python
df_social = pd.read_csv('/Users/konstantinnistratov/PycharmProjects/ml_project_economic_regions/temp/social_df.csv', sep=';')
```

```python
df_economic = pd.read_csv('/Users/konstantinnistratov/PycharmProjects/ml_project_economic_regions/temp/economic_df.csv', sep=';')
```

```python
df_merged = pd.merge(
    df_social, df_economic,
    left_on='regions',
    right_on='regions',
    how='left'
)
```

```python
del df_merged['Unnamed: 0_x']
del df_merged['Unnamed: 0_y']
del df_merged['Unnamed: 15']
del df_merged['Unnamed: 20']
```

```python
df = df_merged.set_index([pd.Index(df_merged['regions'])])
del df['regions']
```

```python
df.sample(50)
```

```python
df.to_csv('/Users/konstantinnistratov/PycharmProjects/ml_project_economic_regions/temp/merged_eco_social.csv', sep=';')
```
