```python
import pandas as pd

pd.set_option('max_colwidth', None)
pd.set_option('display.max_columns', 37)
```

```python
base_economic_df = pd.read_csv('/Users/konstantinnistratov/Downloads/data_regions/section_3.csv', sep=';')
base_economic_df.shape
```

```python
base_economic_df_2020 = base_economic_df[base_economic_df.year == 2022]
base_economic_df_2020_regions = base_economic_df_2020[base_economic_df_2020.object_level == 'регион']

base_economic_df_2020_regions = base_economic_df_2020_regions[['object_name', 'indicator_name', 'indicator_unit', 'indicator_value']]
```

```python
base_economic_df_2020_regions['union_indicator'] = base_economic_df_2020_regions['indicator_name'] + ', ' + base_economic_df_2020_regions['indicator_unit']
del base_economic_df_2020_regions['indicator_name']
del base_economic_df_2020_regions['indicator_unit']
```

```python
base_economic_df_2020_regions
```

```python
economic_2020_reg_dict = base_economic_df_2020_regions.to_dict('list')
```

```python
all_regions = tuple(set(economic_2020_reg_dict.get('object_name')))
len(all_regions)

all_indicators = tuple(set(economic_2020_reg_dict.get('union_indicator')))
len(all_indicators)
```

```python
data_df = pd.DataFrame(columns=['regions', *all_indicators])

for region in all_regions:
    temp_df_region = base_economic_df_2020_regions[base_economic_df_2020_regions.object_name == region]

    temp_dict = {'regions': region}
    for indicator in all_indicators:
        temp_df = temp_df_region[temp_df_region.union_indicator == indicator]
        # temp_df = temp_df[indicator]
        
        # temp_dict[indicator] = temp_df.iat[0,1]
        for value in temp_df.to_dict().get('indicator_value').values():
            temp_dict[indicator] = value

    data_df.loc[len(data_df.index)] = temp_dict
        
data_df
```

```python
data_df.to_csv('/Users/konstantinnistratov/PycharmProjects/ml_project_economic_regions/temp/economic_df.csv', sep=';')
```
